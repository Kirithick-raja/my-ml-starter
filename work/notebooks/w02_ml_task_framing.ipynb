## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

I choose my lane as classification. Since classification predicts a categorical outcome or assign an item to a predefined class.

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

In a classification task, you would predict **the specific class or category** an item belongs to. For example, if classifying emails, you would predict if an email is 'spam' or 'not spam'.

The label (e.g., 'spam', 'not spam', 'cat', 'dog') typically comes from **an observed outcome** that has been historically recorded or manually annotated. This means that for a set of training data, the correct class for each item is already known. In some cases, labels can also be derived from a **defined rule**, especially when setting up an initial classification system or when the outcome is directly determined by specific criteria.

## 3. Success metric

For a classification task, a highly defensible single metric is the **F1-score**, which effectively balances a model's precision (how many selected items are relevant) and recall (how many relevant items are selected). An F1-score ranging from **0.80 to 1.0** typically signifies 'good' performance, as it indicates a strong harmony between avoiding false positives and minimizing false negatives, making it particularly useful for assessing overall model effectiveness across various scenarios.

## 4. The unit of analysis, as a real dataframe

*Load your lane's slice and show it: one row = one what?*

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score, classification_report
from sklearn.datasets import make_classification
X, y = make_classification(
    n_samples=1000, n_features=10, n_informative=5, n_redundant=0,
    n_classes=2, random_state=42, weights=[0.8, 0.2]
)
df = pd.DataFrame(X, columns=[f'feature_{i+1}' for i in range(X.shape[1])])
df['churn'] = y
print("One row represents one customer's data, with various features and a 'churn' target variable.")
display(df.head())
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
model = LogisticRegression(random_state=42, solver='liblinear')
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
f1 = f1_score(y_test, y_pred)
print(f"\nModel F1-Score: {f1:.4f}")
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

## 5. Why ML beats a fixed rule here

The pattern for predicting churn in a customer base is often too messy for simple if-statements because it typically involves a combination of many interacting features (e.g., usage patterns, demographic data, historical interactions). These relationships are rarely linear or easily defined by a few hard-coded rules; instead, they are often subtle, non-linear, and require weighing various factors simultaneously. An if-statement approach would quickly become unmanageable, brittle, and unable to adapt to new, unforeseen patterns, whereas an ML model can learn these complex relationships from data and make more robust predictions.
